```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('default')
```

```python
df = pd.read_csv("student_performance.csv")
df.head()
```

```python
df.shape
df.info()
df.isnull().sum()
```

```python
df.describe()
```

```python
plt.figure(figsize=(7,5))
sns.histplot(df['Final_Exam_Marks'], bins=10, kde=True)
plt.title('Distribution of Final Exam Marks')
plt.xlabel('Final Exam Marks')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.boxplot(x='Gender', y='Final_Exam_Marks', data=df)
plt.title('Final Exam Marks by Gender')
plt.xlabel('Gender')
plt.ylabel('Final Exam Marks')
plt.show()
```

```python
dept_avg = df.groupby('Department')['Final_Exam_Marks'].mean()

plt.figure(figsize=(7,5))
dept_avg.plot(kind='bar')
plt.title('Average Final Exam Marks by Department')
plt.xlabel('Department')
plt.ylabel('Average Marks')
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.scatterplot(x='Attendance_%', y='Final_Exam_Marks', data=df)
plt.title('Attendance vs Final Exam Marks')
plt.xlabel('Attendance (%)')
plt.ylabel('Final Exam Marks')
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.scatterplot(x='Study_Hours_Per_Day', y='Final_Exam_Marks', data=df)
plt.title('Study Hours per Day vs Final Exam Marks')
plt.xlabel('Study Hours Per Day')
plt.ylabel('Final Exam Marks')
plt.show()
```

```python
plt.figure(figsize=(7,5))
df['Grade'].value_counts().plot(kind='bar')
plt.title('Grade Distribution')
plt.xlabel('Grade')
plt.ylabel('Count')
plt.show()
```
